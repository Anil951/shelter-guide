```python
df["seller_type"]=df["seller_type"].map({'OWNER':0,'AGENT':1,'BUILDER':2})
df["layout_type"]=df["layout_type"].map({'BHK':0,'RK':1})
df["property_type"]=df["property_type"].map({'Apartment':0,'Studio Apartment':1,'Independent House':2,'Independent Floor':3,'Villa':4,'Penthouse':5})
df["furnish_type"]=df["furnish_type"].map({'Furnished':0,'Semi-Furnished':1,'Unfurnished':2})
df
```

```python
df=df[df['Latitude']!='Not found'].reset_index(drop=True)
df
```

```python
df = df.dropna()
df.drop_duplicates(inplace = True)
df
```

```python
values_to_need=['2 bathrooms', '1 bathrooms', '3 bathrooms', '6 bathrooms',
       '5 bathrooms', '4 bathrooms', '8 bathrooms', '7 bathrooms',
       '9 bathrooms','12 bathrooms']

df = df[df['bathroom'].isin(values_to_need)]

df['bathroom'] = df['bathroom'].str.replace('bathrooms', '').astype(int)
    
df['price']= [i.replace(',','') for i in df['price']]
df['price']=df['price'].astype(float)

# convert latitude and longitude columns to float
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

df['seller_type']=df['seller_type'].astype(int)
df['bedroom']=df['bedroom'].astype(int)
df['layout_type']=df['layout_type'].astype(int)
df['property_type']=df['property_type'].astype(int)
df['furnish_type']=df['furnish_type'].astype(int)

    # Drop the unwanted values from the 'bathroom' column

df
```

```python
value_counts = df['price'].value_counts()

# use boolean indexing to drop rows where the count is between 1 and 10
to_drop = value_counts[(value_counts >= 1) & (value_counts <= 20)].index
df = df[~df['price'].isin(to_drop)]

# print the cleaned DataFrame
df
```

```python
#removing outliers
desc = df['price'].describe()
q1 = desc.loc['25%']
q3 = desc.loc['75%']
iqr = q3-q1 
lower_lim = q1-(1.5*iqr)
upper_lim = q3+(1.5*iqr)
df=df[(df['price']>=lower_lim)&(df['price']<=upper_lim)]
df
```

```python
import numpy as np
def zscore(col):
    return np.abs((col - col.mean()) / col.std())

# calculate Z-score for 'Salary' column
zscore_salary = zscore(df['Latitude'])

# set a threshold for Z-score
threshold = 3

# identify and show outliers
outliers = df[zscore_salary > threshold]
df.drop(outliers.index, inplace=True)

zscore_salary = zscore(df['Longitude'])
outliers = df[zscore_salary > threshold]
df.drop(outliers.index, inplace=True)

df
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import math
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,VotingRegressor
import xgboost as xgb
from sklearn.metrics import r2_score,mean_absolute_error


X = df.drop(['price','locality'],axis=1).values
y = df[['price']].values
y=y.ravel()

# Set max_features to the square root of the number of features
max_features = int(math.sqrt(X.shape[1]))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=84)

# Define models
rf_model = RandomForestRegressor(n_estimators=1000, random_state=84,max_depth=20, min_samples_split=20, min_samples_leaf=10, max_features=max_features)
xgb_model = xgb.XGBRegressor(n_estimators=800, random_state=84)


# Ensemble models
# Combine the models using voting ensemble
ensemble_model1 = VotingRegressor(estimators=[('rf', rf_model), ('gb', xgb_model)])
ensemble_model1.fit(X, y)

y_pred = ensemble_model1.predict(X_test)
y_pred
r2 = r2_score(y_test, y_pred)
print(f"r2:",r2)

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean absolute :", mae)

import joblib
joblib.dump(ensemble_model1,'FINAL_hyderabad_rent.joblib')
```

```python
#(STRING)LOCATION TO LATITUDE AND LONGITUDE


from geopy.geocoders import Nominatim

# Create a Nominatim geocoder object
geolocator = Nominatim(user_agent="my_app")

# Get user input for location
location_str = input("Enter location: ")

# Use geocoder to convert location string to latitude and longitude
location = geolocator.geocode(location_str)

# Check if location was found
if location is not None:
    latitude = location.latitude
    longitude = location.longitude
    print(f"Latitude: {latitude}")
    print(f"Longitude: {longitude}")
else:
    print("Location not found")
```
