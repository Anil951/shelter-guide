```python
#TOOK 2-3 hrs FOR ONE CSV 

import requests
import pandas as pd

def geocode(location, state):
    # send a GET request to the Nominatim API with the location and state as parameters
    response = requests.get(f"https://nominatim.openstreetmap.org/search?q={location},{state}&format=json")

    # extract the latitude and longitude from the response
    if response.ok:
        data = response.json()
        if data:
            lat = data[0]["lat"]
            lon = data[0]["lon"]
            return lat, lon
    return None, None

# create a sample DataFrame with a 'location' column

# apply the geocoding function to each value in the 'location' column using the 'apply()' method
df3[['latitude', 'longitude']] = df3.apply(lambda x: geocode(x['Location'], 'Tamilnadu'), axis=1, result_type='expand')

print(df3)
```
